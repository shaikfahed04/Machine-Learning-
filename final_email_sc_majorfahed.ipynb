```python
x=dataset #initializing the x value as dataset variable which is all columns expect prediction
y=df['Prediction'] #Prediction column
```

```python
#importing the test and train split package
from sklearn.model_selection import train_test_split
```

**Using KNN classifier for the prediction of the data**

```python
#importing KNN package it is a classifier algorithm
from sklearn.neighbors import KNeighborsClassifier
```

```python
#initializing the Model
model=KNeighborsClassifier(n_neighbors=3,weights='distance',metric='minkowski',p=3)
```

```python
#fitting the above model in KNN
model=model.fit(X_train,y_train)
```

```python
#Creating a search_space
search_space={'n_neighbors':k,'weights':w,'metric':mat,'p':p}
```

```python
#printing the search_space
search_space
```

```python
#initializing the base_model variable to KNN
base_model=KNeighborsClassifier()
```

```python
#Initializing the Grid Search
search=GridSearchCV(estimator=base_model,scoring=accuracy_score,param_grid=search_space)
```

```python
#fitting the train data
results=search.fit(X_train,y_train)
```
