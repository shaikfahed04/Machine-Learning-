#Data preprocessing

```python
#importing packages
from sklearn.preprocessing import LabelEncoder
```

**encoding the class column**

```python
lb=LabelEncoder()
```

```python
#transforming the values
df['class']=lb.fit_transform(df['class'])
```

```python
df
```

here we converted iris setosa as 0 iris verti color as 1 and iris virginica as 2

```python
#we convert object to int
df['class'].info()
```



**Training the dataset**

```python
#creating x and y variables
x=df.drop('class',axis=1)
y=df['class']
```

```python
#importing packages
from sklearn.model_selection import train_test_split
```

```python
#creating test and train varibales
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

#classification of the dataset using KNN

```python
#importing knn package
from sklearn.neighbors import KNeighborsClassifier
```

```python
#creating variable
model=KNeighborsClassifier(n_neighbors=3,weights='uniform',metric='minkowski',p=3)
```

```python
#fitting the model
model=model.fit(X_train,y_train)
```

```python
#predicting the model
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
#importing accuracy score r2_score f1_score and recall packages
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import r2_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
classification_report(y_test,y_pred)
```

#optimization of knn model

```python
#importing gridsearchcv
from sklearn.model_selection import GridSearchCV
```

```python
#Initializing the parameters of the KNN model for the optimization purpose
k=list(range(3,22,2))
w=['uniform','distance']
mat=['minkowski']
p=[1,2,3,4]
```

```python
#Creating a search_space
search_space={'n_neighbors':k,'weights':w,'metric':mat,'p':p}
```

```python
#printing the search_space
search_space
```

```python
#initializing the base_model variable to KNN
base_model=KNeighborsClassifier()
```

```python
#Initializing the GridSearch
search=GridSearchCV(estimator=base_model,scoring=accuracy_score,param_grid=search_space)
```

```python
#fitting the train data
results=search.fit(X_train,y_train)
```

```python
#Getting the best score of the optimized model using Grid Search CV
results.best_score_
```

```python
#getting the best parameters for which we can get best score
results.best_params_
```

```python
#using best parameters and creating a variable called model1
model1=KNeighborsClassifier(n_neighbors=3,weights='uniform',metric='minkowski',p=1)
```
